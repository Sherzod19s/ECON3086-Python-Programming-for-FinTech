# Exercise 9a - Basic operations with dataframe

Using the yfinance library to get Microsoft Stock data of past 5 years.

```python
import pandas as pd
import yfinance as yf

# Get MSFT stock data for the last 5 years using history method
msft_ticker = yf.Ticker('MSFT')
msft = msft_ticker.history(period="5y")
```

```python
msft
```

Q1. List the Index

```python
msft.index
```

Q2. List the Columns Name

```python
msft.columns
```

Q3. Filter the dataframe to get the data of 2024 only.

```python
msft_2024 = msft[msft.index.year == 2024]
msft_2024
```

Q4 Filter the dataframe to get the data of Mar 2023.

```python
msft_march_2023 = msft[(msft.index.year == 2023) & (msft.index.month == 3)]
msft_march_2023
```

Q5.
- Create a new column "Intraday Price Change" where the value is Close price - Open price
- Create a new column "Price Change" where the value is Close price - last trading day Close price
- Create a new column "Percentage change" where the value is close price / last trading day close price - 1

```python
msft['Intraday Price Change'] = msft['Close']-msft['Open']
msft['Price Change'] = msft['Close']-msft['Close'].shift(1)
msft['Percentage change'] = msft['Close']/msft['Close'].shift(1)-1
```

Q6. Check whether there were any dividend within the data

```python
msft[msft['Dividends'] > 0]['Dividends']
```

Q7. Remove the column "Dividends" from the dataframe

```python
msft = msft.drop('Dividends', axis=1)
msft
```

Q8. Find the highest stock price ever in this period, and find the date of it

```python
msft.loc[msft['Close'].idxmax()]
```

Q9. Find the highest volume date

```python
msft.loc[msft['Volume'].idxmax()]
```

Q10. Find the top 20 highest price percentage change date and it's percentage change, order in decending order.

```python
msft_20 = msft['Percentage change'].sort_values(ascending=False).head(20)
msft_20
```
